* Menu de problemas


1. Volumen del cubo
2. Calculando cambio
3. Calculando cambio con IF
4. Calcular Volumen Cilindro
5. Area Triangulo
6. Salir


```python
# Menu de problemas
def VerMenu():
  print("1. Volumen del cubo")
  print("2. Calculando cambio")
  print("3. Calculando cambio con IF")
  print("4. Calcular Volumen Cilindro")
  print("5. Area Triangulo")
  print("6. Salir")
  opcion=int(input("ingrese su opcion:"))
  return opcion

def VolumenCubo():
  A=float(input("Ingrese el valor de A:"))
  B=float(input("Ingrese el valor de B:"))
  C=float(input("Ingrese el valor de C:"))
  V=A*B*C
  print("Calculando Volumen")
  print("El volumen del cubo es:",V)

def CalculandoCambio():
  CP = float(input("Ingrese el costo del producto:"))
  CV = float(input("Ingrese el dinero dado:"))
  Cambio = CV - CP
  print("El cambio es:",Cambio)

def CalculandoCambioConIF():
  CP = float(input("Ingrese el costo del producto:"))
  CV = float(input("Ingrese el dinero dado:"))
  if CV >= CP:
    Cambio = CV - CP
    print("El cambio es:",Cambio)
  else:
    print("El dinero dado es insuficiente.")

def CalcularVolumenCilindro():
  R=float(input("Ingrese el valor del radio:"))
  H=float(input("Ingrese el valor de la altura:"))
  V=3.1416*R*R*H
  print("Calculando Volumen")
  print("El volumen del cilindro es:",V)

def AreaTriangulo():
  B=float(input("Ingrese el valor de la base:"))
  H=float(input("Ingrese el valor de la altura:"))
  A=(B*H)/2
  print("El area del triangulo es:",A)

while True:
  op = VerMenu()
  if op == 1:
    VolumenCubo()
  elif op == 2:
    CalculandoCambio()
  elif op == 3:
    CalculandoCambioConIF()
  elif op == 4:
    CalcularVolumenCilindro()
  elif op == 5:
    AreaTriangulo()
  elif op == 6:
    print("Gracias por usar el programa")
    break
  else:
    print("Opcion no valida")


print("Programa realizado por KARLA SAENZ")
```
